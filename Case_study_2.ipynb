```python
# find the frequency of education column
census['education'].value_counts()
```

```python
#Rename the columns name
census.rename(columns = {'workclass':'employment_type'},inplace = 'True')
census.head()
```

```python
census.rename(columns = {'hours.per.week':'hours_worked'},inplace = 'True')
```

```python
census.head()
```

We have renamed our columns name sucessfully.

```python
# find the frequency of relationship column
census['relationship'].value_counts()
```

```python
# Only find the records of Unamrried people
Unmarried = census[census['relationship']=='Unmarried']
```

```python
# Only find the records of Divorced people
Divorced = census[census['marital.status'] == 'Divorced']
Divorced.head()
```

```python
#Only find those records where people are unmarried and the marital status is divorce

Divorced_1 = Unmarried[Unmarried['marital.status']=='Divorced']
Divorced_1
```

```python
# find the value whose age greater than 50 and sex is male.
old_men = census[(census['age']>50) & (census['sex']=='Male')]
old_men
```

So there are 4568 records shows whoese age more than 50 and they are male only.

```python
# Now we will find how many peoples have completed masters and doing private job
master_private = census[(census['education'] == 'Masters') & (census['employment_type'] == 'Private')]
master_private
```
